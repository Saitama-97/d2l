# 2.1 数据操作

## 2.1.1 pytorch

```python
# 导入torch
import torch

# 具有一个轴的张量对应数学中的向量
# 具有两个轴的张量对应数学中的矩阵

# 创建一个张量（行向量）,其中有12个元素
x = torch.arange(12)
x
```

```python
# 可以通过张量的shape属性查看张量的形状
x.shape
```

```python
# 如果只想知道张量的元素总数(形状的所有元素乘积)
x.numel()
```

```python
# 可以通过reshape改变张量的形状
y = torch.arange(12).reshape(3,4)
y
```

```python
y.shape
```

```python
# 可以不完全指定张量的形状，用-1要求自动计算
z1 = torch.arange(12).reshape(-1,6)  # ?行6列
z2 = torch.arange(12).reshape(3,-1)  # 3行?列
print(z1)
print(z2)
```

```python
# 使用全0、全1、其他常量，快速初始化
t1 = torch.zeros(3,4)
t2 = torch.ones(3,4)
print(t1)
print(t2)
```

```python
# 随机初始化
t3 = torch.randn(3,4)
print(t3)
```

```python
# 使用Python list来初始化
t3 = torch.tensor([
    [2,1,3,4],
    [1,3,4,2],
    [3,4,2,1]
])
print(t3)
```
