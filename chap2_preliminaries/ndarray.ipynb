## 2.1.2 运算符

- 按元素（elementwise）运算
	- 对于任意具有相同形状的张量， 常见的标准算术运算符（+、-、*、/和**）都可以被升级为按元素运算

```python
x = torch.tensor([1.0, 2, 4, 8])
y = torch.tensor([2, 2, 2, 2])
x + y, x - y, x * y, x / y, x ** y, torch.exp(x)  # **运算符是求幂运算 / exp为求幂
```

- 连结（concatenate）操作
	- 沿行（轴-0，形状的第一个元素）
    - 沿列（轴-1，形状的第二个元素）

```python
X = torch.arange(12, dtype=torch.float32).reshape(3, 4)
Y = torch.tensor([
    [2.0, 1, 4, 3],
    [1, 3, 4, 2],
    [3, 4, 2, 1]
])
print(X)
print(Y)
print(torch.cat((X, Y), dim=0))  # 按行连结
print(torch.cat((X, Y), dim=1))  # 按列连结
```

```python
# 判断两个张量是否对应位置相等
X == Y
```

```python
# 对张量的所有元素进行求和
X.sum()
```

## 2.1.3 广播机制

- 正如上面所说，如果两个张量的形状相同，运算是按元素操作（elementwise）
- 如果两个张量的形状不同，也可以进行计算，不过是使用**广播机制**
	- 通过适当复制元素来扩展一个或两个数组，以便在转换之后，两个张量具有相同的形状
    - 对生成的数组执行按元素操作

```python
# 通常是沿着数组中长度为 1 的轴进行广播
a = torch.arange(3).reshape(3, 1)
b = torch.arange(3).reshape(1, 3)
a, b, a + b
```

## 2.1.5 节省内存

- 运行一些操作可能会导致为新结果分配内存。
	- 例如，如果我们用Y = X + Y，我们将取消引用Y指向的张量，而是指向新分配的内存处的张量。

```python
# 举例
before = id(Y)
Y = Y + X
id(Y) == before  # 可以看到这里分配了新的内存地址
```

- 使用切片来执行原地操作

```python
# 对比上面的结果
before = id(Y)
Y[:] = Y + X
id(Y) == before  # 可以看到这里并没有分配新的内存地址
```

```python
# 另一种方法
before = id(Y)
Y += X
id(Y) == before
```

## 2.1.6 转换为其他Python对象

```python
t = torch.tensor([1,2,3,4])
A = t.numpy()  # 将 torch 张量转换成 numpy 数组
B = torch.tensor(A)  # 将 numpy 数组转换成 torch 张量
type(A), type(B)
```

- torch 张量和 numpy 数组**共享底层内存**，即修改一个张量，也会同时修改另一个张量（见下面示例）

```python
C = torch.tensor([2,2,2,2])
B += C
A, t, B, C
```

- 将 torch 张量转换成 Python 标量（item函数或Python的内置函数）

```python
a = torch.tensor([3.5])
a, a.item(), float(a), int(a)
```
